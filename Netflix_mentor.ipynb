# 4 )  Exploratory Data Analysis

```python
# count number of movie and tv show 
type_count = data['type'].value_counts()
type_count
```

```python
# Distribution of content by type using  bar \ pie plot 
fig, axes = plt.subplots(1,2,figsize = (8,4))
sns.countplot (data = data,  x = data['type'] , ax = axes[0] , palette = 'Set2')
plt.pie(type_count , labels = ['Movie' , 'Tv Show'], autopct = '%0.f%%' )
plt.suptitle('Total Content on Netflix' , fontsize = 20 )
# plt.title('Distribution of content by Type')
# plt.ylabel('Number of Count')
plt.show()
```

```python
# most common Genres
data.head()
```

```python
# create new column genres
data['genres'] = data['listed_in'].apply(lambda x : x.split(','))
```

```python
data.head()
```

```python
all_genres = sum(data['genres'], [])
genre_count = pd.Series(all_genres).value_counts().head(10)
top_10 = genre_count.nlargest(10)
bottom_10 = genre_count.nsmallest(10)
print('top 10 \n' , top_10)
print('Bottom 10 \n', bottom_10)
```

```python
plt.figure( figsize = (9,4))
sns.barplot(x = genre_count.index , y = genre_count.values )
plt.title('Most common Genress in Netflix')
plt.ylabel('Number of counts')
plt.xlabel('Genre type')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
data.columns
```

```python
# Rating count on netflix
rating_count = data['rating'].value_counts().reset_index().sort_values(by = 'count' , ascending = False)
rating_count
```

```python
# rating count on Netflix using bar 
plt.figure ( figsize = (8,4))
plt.bar(rating_count['rating']  , rating_count['count'])
plt.xticks(rotation=45, ha='right')
plt.xlabel("Rating Types")
plt.ylabel("Rating Frequency")
plt.suptitle('Rating on Netflix', fontsize=20)
plt.show()
```

```python
# bar for rating on Netflix
plt.pie(  rating_count['count'][:8] ,labels =rating_count['rating'][:8], autopct = '%0.0f%%')
plt.suptitle('Rating on Netflix' , fontsize = 20)
plt.show()
```

```python
# based on country 
data['country'].value_counts()
```

```python
top_country = data['country'].value_counts().reset_index().sort_values (by ='count', ascending = False)[:10]
top_country
```

```python
# top 10 countries base on content on netflix
plt.figure ( figsize = (10,5))
plt.bar(top_country['country'] , top_country['count'])
plt.xticks(rotation = 45 , ha ='right')
plt.suptitle('Top 10 countries base on content on Netflix' , fontsize = 15)
plt.xlabel ('Country Name')
plt.ylabel('Number of Count')
plt.show()
```

```python
# MonthlyreleasesofMoviesandTVshowson Netflix
data['Month'] = data.date_added.dt.month
data['Month'].unique()
```

```python
monthly_movie_release = data[data['type'] == 'Movie']['Month'].value_counts().sort_index()
monthly_Tvshow_release = data[data['type'] == 'TV Show']['Month'].value_counts().sort_index()
```

```python
# MonthlyreleasesofMoviesandTVshowson Netflix
plt.figure(figsize=(8,4))
plt.plot(monthly_movie_release.index , monthly_movie_release.values , label = 'Movie')
plt.plot(monthly_Tvshow_release.index , monthly_Tvshow_release.values , label = 'Tv Show')
plt.ylabel('Frequency of Release')
plt.xlabel('Month')
plt.legend()
plt.grid(True)
plt.xticks(range(1,13) , ['JAN', 'FEB','MAR','APR','MAY','JUN','JULY','AUG','SEP','OCT','NOV','DEC'])
plt.suptitle('Monthly releases of Movies and TVshowson Netflix' , fontsize = 17)
plt.show()
```

```python
# yearly  releases of Movies and TVshowson Netflix
data['add_year'] = data['date_added'].dt.year
movie_year = data[data['type'] == 'Movie']['add_year'].value_counts().sort_index(ascending = False)
Tv_show_year = data[data['type'] == 'TV Show']['add_year'].value_counts().sort_index(ascending = False)
```

```python
# yearly  releases of Movies and TVshows on Netflix
plt.figure(figsize = (8,4))
plt.plot(movie_year.index ,movie_year.values , label = 'Movie') 
plt.plot(Tv_show_year.index , Tv_show_year.values , label = 'Tv show')
plt.xlabel('Year')
plt.ylabel('Frequency of releasese')
plt.grid(True)
plt.legend()
plt.suptitle('yearly  releases of Movies and TV shows on Netflix' , fontsize = 17)
plt.show()
```

```python
# Top 10 popular genres for movies on Netflix
movie_genres = data[data['type'] =='Movie'].groupby('listed_in').size().sort_values(ascending = False)[:10]
plt.figure(figsize = (10,5))
plt.bar(movie_genres.index , movie_genres.values )
plt.xticks(rotation = 45 , ha = 'right')
plt.xlabel('Genre')
plt.ylabel('Movie Frequency')
plt.suptitle('Top 10 popular genres for movies on Netflix' , fontsize = 17)
plt.show()
```

```python
# Top 10 popular genres for Tv show on Netflix
tvshow_genres = data[data['type'] =='TV Show'].groupby('listed_in').size().sort_values(ascending = False)[:10]
plt.figure(figsize = (10,5))
plt.bar(tvshow_genres.index , tvshow_genres.values )
plt.xticks(rotation = 45 , ha = 'right')
plt.xlabel('Genre')
plt.ylabel('Tv Show Frequency')
plt.suptitle('Top 10 popular genres for Tv Show on Netflix' , fontsize = 17)
plt.show()
```

```python
# Top 15 directors across Netflix with hoigh frequency of movies and Tv shows
top_directors = data['director'].value_counts().reset_index().sort_values(by = 'count', ascending = False)[1:16]
top_directors
```

```python
plt.figure(figsize = (10,5))
plt.bar(top_directors['director'] , top_directors['count'])
plt.xticks(rotation = 45, ha ='right')
plt.xlabel('Director')
plt.ylabel('Frequency of movie and Tv show')
plt.suptitle('Top 15 directors across Netflix with hoigh frequency of movies and Tv shows', fontsize = 15)
plt.show()
```
